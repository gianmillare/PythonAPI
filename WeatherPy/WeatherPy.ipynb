```python
# Dependencies
import os
import csv
import json
import requests

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from pprint import pprint
from apikey import apikey
from citipy import citipy

latrange = (-90, 90)
lngrange = (-180, 180)
```

```python
# Given Information
lat_lngs = []
cities = []

lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    if city not in cities:
        cities.append(city)

len(cities)
```

```python
# Set quantifiable data equal to Zero
record = 0
recordset = 1

latitude = []
longitude = []
temperature = []
cloudiness = []
humidity = []
windspeed = []
date = []
cityxyz = []
country = []
```

```python
# Testing URL
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + apikey
url
```

```python
# Create For Loop
print('---------------------------------------------------------------------------------')

url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + apikey 

for city in cities:
    
    query_url = url + "&q=" + city
    responsejson = requests.get(query_url).json()
    
    if record < 50:
        record = record + 1
    else:
        recordset = recordset + 1
        record = 0

    print('Processing record ' + str(record) + ' of set ' + str(recordset) + ' -----> ' + city)

    try:
        latitude.append(responsejson['coord']['lat'])
        longitude.append(responsejson['coord']['lon'])
        temperature.append(responsejson['main']['temp_max'])
        humidity.append(responsejson['main']['humidity'])
        cloudiness.append(responsejson['clouds']['all'])
        windspeed.append(responsejson['wind']['speed'])
        date.append(responsejson['dt'])
        country.append(responsejson['sys']['country'])
        cityxyz.append(city)

    except:
        print("City not found. Skipping...")
        pass

print('---------------------------------------------------------------------------------')
```

```python
# Create Data Frame
df_prep = {"City": cityxyz, "Cloudiness" : cloudiness, "Country" : country,
           "Date" : date, "Humidity" : humidity, "Temp": temperature, 
           "Lat" : latitude, "Lng" : longitude, "Wind Speed" : windspeed}
dataframe = pd.DataFrame(df_prep)
```
